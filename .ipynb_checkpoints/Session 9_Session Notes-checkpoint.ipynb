<h2>Line Chart</h2>

```python
# Line chart
df_apple = pd.read_csv('Data/AAPL.csv', parse_dates=['Date'], index_col=['Date'])
df_apple.head(10)

df_apple = pd.DataFrame(df_apple.High)

df_apple_daily = df_apple.asfreq('D', method='ffill')

df_apple_daily['pct_change_daily'] = df_apple_daily.High.pct_change().mul(100)
df_apple_daily['pct_change_monthly'] = df_apple_daily.High.pct_change(30).mul(100)
df_apple_daily['pct_change_yearly'] = df_apple_daily.High.pct_change(360).mul(100)

df_apple_daily.loc[:,['pct_change_daily','pct_change_monthly', 'pct_change_yearly']].plot()

plt.show()

```

<h2>Histogram</h2>

```python
from datetime import date, timedelta

df_customer_birthdays = pd.read_csv('Data/customer_birthdates.csv',  parse_dates=['BirthDate'])

df_customer_birthdays['Today'] = pd.Timestamp.today()
df_customer_birthdays['Age_Years'] =  df_customer_birthdays.Today.sub(df_customer_birthdays.BirthDate).div(timedelta(days=365.2425))

df_customer_birthdays.Age_Years = df_customer_birthdays.Age_Years.astype(int)

df_customer_birthdays.head()
```

```python

#Describe()
df_customer_birthdays.describe()
```

```python
#Plot histogram - method 1

plt.hist(df_customer_birthdays.Age_Years)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Customer Age Histogram')
plt.show()
```

```python
#Plot histogram - method 2

df_customer_birthdays.Age_Years.hist()
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Customer Age Histogram')
plt.show()
```

<h2>Bar Chart</h2>


```python
df_customer_orders = pd.read_csv('Data/customers_orders.csv')
df_customer_orders_us = df_customer_orders[df_customer_orders.COUNTRY_CODE == 'US']
```

```python
#Get a count of each state  -- method 1 (use size())

df_by_state = pd.DataFrame(df_customer_orders_us[['STATE']].groupby('STATE').size(), columns=['Count'])
states = df_by_state.index
N = len(states)
x = range(N)
plt.clf()
plt.bar(x, df_by_state.Count)
plt.xticks(x, states, rotation=90)
plt.rcParams["figure.figsize"] = [10,100]
plt.show()
```

```python
#Get a count of each state  -- method 1 (use size())

df_by_state = pd.DataFrame(df_customer_orders_us[['STATE']].groupby('STATE').size(), columns=['Count'])

df_by_state
```

```python
#Get a count of each state  -- method 2 (use aggregate functions)

df_customer_orders_us.groupby(['STATE'])['STATE'].agg('count')
```

<h2>Aggregating values</h2>
Summing columns/rows

```python
# How to create "Total" columns/ "Total Row"
# Case 1: 

df_sales = pd.DataFrame({'Region':['Soutwest', 'Central', 'Northeast'] ,
                         'Jan': [10,20,30], 'Feb': [29,39,49], 
                         'Comments':['Jan was below average','Marc was above average','Jan was good'], 
                         'Mar':[50,90,10]})

df_sales.set_index('Region', inplace=True)

df_sales
```

```python
# create df_sales['Total'] column
df_sales['Total'] = df_sales[['Jan','Feb','Mar']].sum(axis=1)

df_sales

```

```python
# create df_sales Total Row:

df_sales.loc['Total', 'Jan'] = df_sales.Jan.sum()

df_sales.loc['Total', 'Feb'] = df_sales.Feb.sum()

df_sales.loc['Total', 'Mar'] = df_sales.Mar.sum()

df_sales
```

```python
# Now we can plot the totals using a bar chart ...

months = ['Jan', 'Feb', 'Mar']
N = len(months)
x = np.arange(N)

plt.bar(x, df_sales.loc['Total', ['Jan', 'Feb', 'Mar']].values)
plt.title('Bar Char')
plt.xlabel('Value')
plt.ylabel('Y Value')
plt.xticks(x, months)

plt.show()
```
