<h2>Question 2</h2>

Using the midterm data set (customer_orders.csv) - you want want to create a histogram of the gender distribution for the customer orders.  Your histogram should show only two bars - one for Male (Mr. and Master.) and Female (Mrs. and Miss.).

```python
df = pd.read_csv('Data/customers_orders.csv')
```

```python
df.GenderCode.unique()
```

```python
df['Gender'] = 'Male'
```

```python
df.loc[df.GenderCode == 'Mrs.', ['GenderCode', 'Gender']] 
```

```python
df.loc[df.GenderCode == 'Mrs.', ['Gender']] = 'Female' 
```

```python
df.loc[df.GenderCode == 'Miss.', ['Gender']] = 'Female'
```

```python
df.Gender.unique()
```

```python
plt.hist(df.Gender)

plt.title('Male vs Female')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
