<h3>Question 3: Can you regress how delayed a flight will be before it is delayed? </h3>
<h2><u>Modeling Delays</u></h2>

```r
library(biganalytics)

```

```r

fx<-ArrDelay ~ Year+Month+UniqueCarrier+DayOfWeek+DepTime+DayofMonth+ActualElapsedTime+DepDelay
```

```r
blm <- biglm.big.matrix(fx, data=x)
summary(blm)
```

```r
fx2<-DepDelay ~ Year+Month+UniqueCarrier+DayOfWeek+DepTime+DayofMonth+ActualElapsedTime
```

```r
blm <- biglm.big.matrix(fx2, data=x)
summary(blm)
```

Based on the above regression equation, the day of the week is the most imporant factor 



<h2><u>Adding new Variable</u></h2>

We assigned the extra column named "Age" to a new calculated variable of the difference between DepDelay and ArrDelay. This variable is meant to measure whether delayed flights try to increase speed or take a faster route to try and recover the lost time.

```r
x[,"age"] <- x[,"DepDelay"] -  x[,"ArrDelay"]
```

```r
fx3<-DepDelay ~ Year+Month+UniqueCarrier+DayOfWeek+DepTime+DayofMonth+ActualElapsedTime+age
```

```r
blm <- biglm.big.matrix(fx3, data=x)
summary(blm)
```
