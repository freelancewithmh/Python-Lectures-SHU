```python
#Merge

df_merge = df[['CUSTNAME' , 'GenderCode' , 'ADDRESS1' , 'CITY' , 'STATE' , 'COUNTRY_CODE' , 'POSTAL_CODE' , 'POSTAL_CODE_PLUS4' , 'ADDRESS2', 'EMAIL_ADDRESS' , 'PHONE_NUMBER' , 'CREDITCARD_TYPE' , 'SALESMAN_ID' , 'LOCALITY' , 'NATIONALITY' , 'NATIONAL_ID' , 'CREDITCARD_NUMBER' , 'DRIVER_LICENSE' , 'CUST_ID' , 'ORDER_ID' , 'ORDER_DATE' , 'ORDER_TIME' , 'FREIGHT_CHARGES' , 'ORDER_SALESMAN' , 'ORDER_POSTED_DATE' , 'ORDER_SHIP_DATE' , 'ORDER_VALUE' , 'T_TYPE', 'PURCHASE_TOUCHPOINT' , 'PURCHASE_STATUS' , 'ORDER_TYPE' , 'GENERATION']]
```

```python
pd.merge(df_merge , df_customerbirthdate, how='left', left_on = df_merge.EMAIL_ADDRESS , right_on = df_customerbirthdate.EMAIL_ADDRESS )
```

```python
df_combined_merge = pd.merge(df_merge , df_customerbirthdate, how='left', left_on = df_merge.EMAIL_ADDRESS , right_on = df_customerbirthdate.EMAIL_ADDRESS )
```
